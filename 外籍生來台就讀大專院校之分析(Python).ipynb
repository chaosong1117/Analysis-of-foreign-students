# 106~110年來自亞洲國家之歷屆學位生_正式修讀學位外國生國家前10排行

```python
pd.DataFrame(df.loc[df['洲別'] == '亞洲'].groupby(['年', '國別']).sum()\
.sort_values(by='學位生_正式修讀學位外國生', ascending=False)['學位生_正式修讀學位外國生'].head(10))
```

```python
df.loc[:, '學位生_總和'] = df['學位生_正式修讀學位外國生'] + df['學位生_僑生.含港澳.'] + df['學位生_正式修讀學位陸生']
```

```python
df.loc[:, '非學位生_總和'] = df['非學位生_外國交換生'] + df['非學位生_外國短期研習及個人選讀'] + df['非學位生_大專附設華語文中心學生']\
+ df['非學位生_大陸研修生'] + df['非學位生_海青班']
```

```python
df
```

```python
df.groupby(['年', '洲別'])['學位生_總和', '非學位生_總和'].sum()
```

```python
df5 = df[(df['年'] == 110) & (df['洲別'] == '亞洲')].set_index('年')
df5
```

# 110年度大陸地區佔全亞洲學位生之比例

```python
pd.DataFrame(100 * (df5[df5['國別'] == '大陸地區']['學位生_總和'] / sum(df5['學位生_總和'])))
```

# 110年度各亞洲國家佔全亞洲國家學位生之比例前五名的國家

```python
df7 = df5.groupby('國別').sum()
df7.sort_values(by='學位生_總和', ascending=False, inplace=True)
```

```python
pd.DataFrame(df7['學位生_總和'].head())
```

```python
df7['學生位總和百分比'] = 100 * (df7['學位生_總和'] / df7['學位生_總和'].sum())
df7
```

```python
df7['非學生位總和百分比'] = 100 * (df7['非學位生_總和'] / df7['非學位生_總和'].sum())
df7
```
